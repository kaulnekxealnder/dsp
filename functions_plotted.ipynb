```python
import matplotlib
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x): # sample function # type in your function here
    return x **3 - x**2 # x squared
```

```python
n1 = -5.0# assign
n2 =  5.0#assign
nslice = 50# assign
x = np.linspace(n1, n2, nslice) # defined linear space, num equal spaces
y = f(x) # y is this function we are calling f(x) and assigning to y
```

```python
def draw_tangent(x,y,a): # using x, function y, and a as x proxies
    for a in x: # for each number in the x lin space
        h = 1/1000000 # shift factor for small slicing
        f_not = (f(a + h)-f(a))/h # derivative at all points a in sample
        tan = f(a) + f_not*(x-a) # tangent
        plt.plot(a, f(a), 'om', x,tan, 'r--', alpha=0.01)
        plt.ylim(ymax=10, ymin=-5)
        plt.xlim(xmax=5, xmin=-5)
        
for a in x:
    draw_tangent(x, y, a)

plt.plot(x, y, alpha=0.75)
plt.show()
```
